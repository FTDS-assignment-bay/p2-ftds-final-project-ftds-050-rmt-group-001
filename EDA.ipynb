## EDA

### Distribusi Rating dari Top 5 Model Handphone

```python
top5_models = (
    df[['model_name', 'overall_rating']]
    .sort_values(by='overall_rating', ascending=False)
    .drop_duplicates()
    .head(5)
)

# Barplot
plt.figure(figsize=(12, 6))
ax =sns.barplot(
    data=top5_models,
    x='model_name',
    y='overall_rating',
    palette='viridis',
    order=top5_models['model_name']
)

plt.title('5 Model Ponsel dengan Overall Rating Tertinggi')
plt.xlabel('Model Handphone')
plt.ylabel('Overall Rating')
plt.xticks(rotation=30)

for p in ax.patches:
    ax.annotate(
        f'{p.get_height():.2f}',
        (p.get_x() + p.get_width() / 2., p.get_height()),
        ha='center',
        va='bottom',
        fontsize=10,
        xytext=(0, 5),
        textcoords='offset points'
    )

plt.show()
```

Visualisasi menunjukkan bahwa Galaxy S25 FE memiliki overall rating tertinggi yaitu sebesar 4.60, menandakan tingkat kepuasan pengguna paling baik di antara lima model teratas. Galaxy Z Flip7 FE dan Galaxy Z Fold7 berada tepat di bawahnya dengan rating yang sama yaitu 4.50, mengindikasikan bahwa meskipun berada pada segmen dan desain berbeda, kualitas pengalaman pengguna dinilai setara.Di tutup dengan,model Samsung S22+ dan Samsung Galaxy S23 FE dengan masing-masing rating keseluruhan 4.40 dan 4.30, yang tetap tergolong sangat baik. Secara keseluruhan, selisih rating antar model relatif kecil, menunjukkan bahwa kelima ponsel ini sama-sama berada di kelas performa tinggi.

### Distribusi Harga Handphone Samsung

```python
#Harga untuk Masing-Masing Model
price_per_model = df[['model_name', 'price']].drop_duplicates()

# Histogram Harga
plt.figure(figsize=(8, 6))
sns.histplot(
    price_per_model['price'],
    bins=30,
    kde=True
)

plt.title("Distribusi Harga")
plt.xlabel("Harga")
plt.ylabel("Frekuensi")
plt.tight_layout()
plt.show()
```

Histogram distribusi harga per model menunjukkan bahwa sebagian besar model handphone berada pada rentang harga rendah hingga menengah, dengan konsentrasi tertinggi di kisaran sekitar 200–450, sementara jumlah model dengan harga tinggi relatif sedikit. Distribusi harga tampak miring ke kanan (right-skewed), yang mengindikasikan adanya beberapa model berharga sangat tinggi sebagai outlier di sisi kanan grafik. Garis KDE menunjukan bahwa pasar didominasi oleh model dengan harga terjangkau hingga menengah, sedangkan model premium hanya menyumbang sebagian kecil dari keseluruhan data. Pola ini menunjukkan bahwa variasi harga antar model cukup besar, namun mayoritas model difokuskan pada segmen pasar menengah.

### Korelasi Specs Handphone terhadap Harga

```python
numeric_cols = [
    'price',
    'ram',
    'battery',
    'weight_gr',
    'cpu_speed',
    'storage'
]

df_corr = df[numeric_cols]

# Hitung korelasi
corr_matrix = df_corr.corr()

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(
    corr_matrix,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    linewidths=0.5
)

plt.title("Korelasi Spesifikasi Handphone terhadap Harga")
plt.tight_layout()
plt.show()
```

Heatmap korelasi menunjukkan bahwa harga handphone memiliki hubungan paling kuat dengan kecepatan CPU (0,61) dan kapasitas storage (0,57), yang mengindikasikan bahwa performa prosesor dan ruang penyimpanan merupakan salah satu faktor utama dalam penentu harga handphone. RAM memiliki korelasi sedang terhadap harga (0,36), sedangkan   `battery` dan berat `weight_gr` menunjukkan korelasi yang relatif lemah, sehingga kapasitas baterai dan bobot perangkat tidak terlalu berpengaruh langsung pada harga jual. Selain itu, terlihat korelasi negatif lemah antara battery dengan `cpu_speed` (-0,20), yang mengindikasikan bahwa ponsel dengan performa tinggi cenderung tidak selalu memiliki kapasitas baterai lebih besar.

### Hubungan Harga Terhadap Rating

```python
df_harga_rating = df[['model_name', 'price', 'overall_rating']].drop_duplicates()

#Boxplot Hubungan antara Harga dan Rating
plt.figure(figsize=(10, 7))
ax = sns.scatterplot(
    data=df_harga_rating,
    x='overall_rating',
    y='price'
)

# Tambahkan label model di tiap titik
for i in range(len(df)):
    ax.text(
        df['overall_rating'].iloc[i] + 0.01,  
        df['price'].iloc[i],
        df['model_name'].iloc[i],
        fontsize=8,
        alpha=0.7
    )

plt.title("Harga vs Rating")
plt.xlabel("Rating")
plt.ylabel("Harga")
plt.tight_layout()
plt.show()


```

Scatter plot menunjukkan bahwa tidak terdapat korelasi linear yang kuat antara harga dan rating. Ponsel dengan rating di atas 4.3 tersebar pada berbagai rentang harga, mulai dari kelas menengah hingga sangat mahal. Terlihat satu model handphone dengan harga sangat tinggi yaitu model Z Fold 7 dengan harga di atas 2000, namun ratingnya tidak jauh berbeda dibanding handphone dengan harga jauh lebih rendah seperti Galaxy Z Flip7 FE dan Galaxy S25 FE. Hal ini mengindikasikan bahwa harga mahal tidak selalu menjamin rating yang jauh lebih baik, dan kepuasan pengguna lebih dipengaruhi oleh faktor lain seperti keseimbangan fitur, performa, dan value for money dibanding harga semata.

### Specs Comparison 

```python
# Group spesifikasi per model
spec_comparison = (
    df.groupby('model_name')
      .agg({
          'price': 'max',
          'battery': 'max',
          'ram': 'max',
          'storage': 'max',
          'screen_size': 'max',
          'refresh_rate': 'max',
          'weight_gr': 'max'
      })
      .round(2)
      .sort_values(by='price', ascending=False)
)

print("Perbandingan Spesifikasi Handphone:\n")
print(spec_comparison.to_string())
```

### WordCloud 

```python
#Penglabelan Review

df['rating_label'] = np.where(
    df['ratings'].between(4, 5),
    'Positive',
    'Negative'
)
```

#### WordCloud Positive Before PreProcessing

```python
# WordCloud Positive
text_positive = df[df['rating_label'] == 'Positive'].reviews.values
cloud_text_positive = WordCloud(background_color='black',colormap="cool",collocations=False,width=2000,height=1000).generate(" ".join(text_positive))

# Menampilkan wordcloud
plt.figure(figsize=(15,10))
plt.axis('off')
plt.title("Positive Review Words",fontsize=20)
plt.imshow(cloud_text_positive)
```

WordCloud menunjukkan bahwa kata “phone” dan “screen” paling dominan, menandakan bahwa perangkat secara keseluruhan dan kualitas layar merupakan aspek yang paling sering diapresiasi pengguna. Kata-kata seperti “good”, “great”, “excellent”, dan “love” memperkuat sentimen kepuasan yang tinggi. Selain itu, kemunculan kata “battery” dan “camera”mengindikasikan bahwa daya tahan baterai dan kualitas kamera sebagai faktor penambah yang memengaruhi review positif. Secara keseluruhan, wordcloud ini menunjukkan bahwa kombinasi kualitas layar, performa, dan fitur utama menjadi alasan utama tingginya kepuasan pengguna terhadap ponsel Samsung.

#### WordCloud Negative Before Preprocessing

```python
# WordCloud Negative
text_negative =  df[df['rating_label'] == 'Negative'].reviews.values
cloud_text_negative = WordCloud(background_color='black',colormap="flare",collocations=False,width=2000,height=1000).generate(" ".join(text_negative))

# Menampilkan wordcloud
plt.figure(figsize=(15,10))
plt.axis('off')
plt.title("Negative Review Words",fontsize=20)
plt.imshow(cloud_text_negative)
```

WordCloud negatif didominasi oleh kata 'phone', 'screen', dan 'time', yang menunjukan bahwa masalah paling sering berkaitan dengan fungsi perangkat dan pengalaman penggunaan dalam jangka waktu tertentu. Kata-kata seperti 'issue', 'problem', 'locked', dan 'unlock' menandakan adanya keluhan terkait sistem penguncian perangkat, atau masalah keamanan. Kemunculan kata 'Amazon', 'seller', 'return', dan 'refund' menunjukkan bahwa sebagian besar review negatif juga dipengaruhi oleh pengalaman pasca-pembelian, seperti pengembalian barang, layanan penjual, atau kondisi produk (misalnya "refurbished"). Selain itu, kata 'battery', 'camera', dan 'network' mengindikasikan keluhan pada performa perangkat. Secara keseluruhan, insight ini menunjukkan bahwa review negatif tidak hanya disebabkan oleh spesifikasi teknis, tetapi juga oleh kualitas unit dan layanan distribusi.

## EDA Conclusion

Berdasarkan seluruh proses Exploratory Data Analysis (EDA) yang telah dilakukan, dapat disimpulkan bahwa **rating ponsel Samsung** tidak sepenuhnya ditentukan oleh **harga**, melainkan oleh **keseimbangan antara spesifikasi, performa, dan pengalaman pengguna**. Analisis hubungan harga dan rating menunjukkan tidak adanya korelasi linear yang kuat, di mana beberapa model dengan harga menengah mampu memperoleh rating setara atau bahkan lebih tinggi dibanding model berharga mahal.

Dari sisi spesifikasi, fitur seperti **RAM, storage, dan CPU speed** memiliki hubungan yang lebih relevan terhadap harga dibanding baterai atau berat perangkat, sehingga dapat menjadi faktor penting dalam penentuan nilai jual. Analisis sentimen melalui WordCloud memperkuat temuan ini, di mana **review positif didominasi oleh apresiasi terhadap kualitas layar, performa, kamera, dan daya tahan baterai**, sedangkan **review negatif lebih banyak dipicu oleh isu teknis, penguncian perangkat, serta pengalaman layanan penjual dan proses pengembalian**.

Secara keseluruhan, EDA ini menunjukkan bahwa **nilai terhadap harga** (value for money) dan **kualitas pengalaman pengguna** menjadi faktor utama dalam membentuk persepsi konsumen. Insight ini dapat digunakan sebagai dasar pemilihan fitur pada tahap pemodelan, pengembangan sistem rekomendasi, serta pengambilan keputusan bisnis terkait strategi produk dan distribusi.
